```python
from typing_extensions import Self
from ast import If
import random

# ìºë¦­í„° í´ë˜ìŠ¤
class Character:
  def __init__(self, name, level=1, health=100, attack=25, defense=5):
    self.name = name
    self.level = level
    self.health = health
    self.attack = attack
    self.defense = defense

    # ì²´ë ¥
  def is_alive(self):
      return self.health > 0

    # ê³µê²© ë‹¹í•˜ê¸°
  def take_damage(self, damage):
    actual_damage = damage - self.defense
    if actual_damage < 0:
      actual_damage = 0
    self.health -= actual_damage

    # ê³µê²© í•˜ê¸°
  def attack_target(self, target):
      damage = random.randint(1, self.attack)
      print(f"{self.name}, {target.name} ê³µê²©í•˜ì—¬ {damage}ì˜ ë°ë¯¸ì§€")
      target.take_damage(damage)

# Player í´ë˜ìŠ¤
class Player(Character):
  def __init__(self, name):
    Character.__init__(self, name, level=1, health=100, attack=25, defense=5)
    self.experience = 0 # ê²½í—˜ì¹˜

  # ê²½í—˜ì¹˜ ë©”ì„œë“œ
  def gain_experience(self, amount):
    self.experience += amount
    print(f"{self.name}, {amount}ì˜ ê²½í—˜ì¹˜ íšë“!")

# Monster í´ë˜ìŠ¤
class Monster(Character):
  def __init__(self, name, level):
    health = random.randint(10, 30) * level
    attack = random.randint(5, 20) * level
    defense = random.randint(1, 5) * level
    Character.__init__(self, name, level, health, attack, defense)

# battle í•¨ìˆ˜
def battle(player, monster):
    while player.is_alive() and monster.is_alive():
      # player ê³µê²©
      damage = random.randint(1, player.attack) - monster.defense
      if damage > 0:
        monster.health -= damage
        print(f"{player.name}, {monster.name}ì—ê²Œ {damage}ë§Œí¼ ê³µê²©!")
        print(f"{monster.name}ì˜ ì²´ë ¥: {monster.health}")

      else:
        print(f"{monster.name}ì—ê²Œ íš¨ê³¼ê°€ ì—†ëŠ”ë“¯ í•˜ë‹¤...")

      # monster ê³µê²©
      damage = random.randint(1, monster.attack) - player.defense
      if damage > 0:
        player.health -= damage
        print(f"{monster.name}, {player.name}ì—ê²Œ {damage}ë§Œí¼ ê³µê²©!")
        print(f"{player.name}ì˜ ì²´ë ¥: {player.health}")
      else:
        print(f"{player.name}ì—ê²Œ íš¨ê³¼ëŠ” ë¯¸ë¯¸í–ˆë‹¤!")

    if player.is_alive():
      print(f"íœ´... {monster.name}ë¥¼ ì²˜ì¹˜í–ˆë‹¤!")
      player.gain_experience(monster.level * 20)
    else:
      print("ì¡Œë‹¤...")
      return False
    return True

# game over í›„ ë‹¤ì‹œ ì‹œì‘í• ì§€ ë¬»ëŠ” í•¨ìˆ˜
def play_again():
  while True:
    answer = input("ë‹¤ì‹œ ì‹œì‘í• ê¹Œìš”? (y/n): ")
    if answer == "y":
      return True
    elif answer == "n":
      print("GAME OVER")
      return False
    else:
      print("ì˜ëª»ëœ ì…ë ¥ì…ë‹ˆë‹¤.")

# ë°ì½”ë ˆì´í„° í•¨ìˆ˜
def game_victory_decorator(func):
    def wrapper(*args, **kwargs):
        print("â•”â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•—")
        print("â•‘           ğŸ® GAME CLEAR! ğŸ®            â•‘")
        print("â• â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•£")
        func(*args, **kwargs)
        print("â• â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•£")
        print("â•‘       ğŸ‘‘ Victory is Yours! ğŸ‘‘         â•‘")
        print("â•šâ•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•â•")
    return wrapper
# ìŠ¹ë¦¬ ë©”ì‹œì§€ ì¶œë ¥ í•¨ìˆ˜
@game_victory_decorator
def victory_message():
    print("ëª¨ë“  ëª¬ìŠ¤í„°ë¥¼ ë¬¼ë¦¬ì³¤ìŠµë‹ˆë‹¤. ìŠ¹ë¦¬!")

# main í•¨ìˆ˜
def main():
    monster_dict = {'ìŠ¬ë¼ì„': 1, 'ê³ ë¸”ë¦°': 2, 'ì˜¤í¬': 3}  # ëª¬ìŠ¤í„° ì¢…ë¥˜

    while True:  # ê²Œì„ ë£¨í”„
        player_name = input("í”Œë ˆì´ì–´ì˜ ì´ë¦„ì„ ì…ë ¥í•˜ì„¸ìš”: ")  # ì´ë¦„ ì…ë ¥
        player = Player(player_name)

        for monster_name, level in monster_dict.items():
            print(f"\n{monster_name}ì™€ ì „íˆ¬ ì‹œì‘!")
            monster = Monster(monster_name, level)
            '''ì´ ë°‘ì— ë£¨í”„ ë¶€ë¶„ì€ ì¸í„°ë„·ì˜ ë„ì›€ì„ ì¢€ ë°›ì•˜ìŠµë‹ˆë‹¤'''
            if not battle(player, monster):  # ì „íˆ¬ì—ì„œ íŒ¨ë°°í–ˆì„ ê²½ìš°
                while True:  # ì¬ì‹œì‘ ì—¬ë¶€ ë°˜ë³µí•´ì„œ ë¬»ê¸°
                    if play_again():  # ì¬ì‹œì‘ ì„ íƒ
                        break  # ë‚´ë¶€ ë£¨í”„ ì¢…ë£Œ, ì™¸ë¶€ ê²Œì„ ë£¨í”„ ì¬ì‹œì‘
                    else:  # ê²Œì„ ì¢…ë£Œ ì„ íƒ
                        return
                break  # ì™¸ë¶€ ë£¨í”„ì—ì„œ ë‹¤ìŒ ê²Œì„ìœ¼ë¡œ ì¬ì‹œì‘

        # ëª¨ë“  ëª¬ìŠ¤í„°ë¥¼ ì²˜ì¹˜í•œ ê²½ìš°
        else:  # `for` ë£¨í”„ê°€ ì •ìƒì ìœ¼ë¡œ ëë‚¬ì„ ë•Œ ì‹¤í–‰
            victory_message()  # ìŠ¹ë¦¬ ë©”ì‹œì§€ ì¶œë ¥
            return  # ê²Œì„ ì¢…ë£Œ


# ê²Œì„ ì‹¤í–‰
main()

```
